```python
# Load data
events_oct = spark.read.csv("/Volumes/workspace/ecommerce_idc/ecommerce_data_idc/2019-Oct.csv",header=True,inferSchema=True)

# This is Lazy Evaluation, As of now spark doesn't read the data, its just create the data frame and infer the schema
```

### Basic Operations

```python
display(events_oct.limit(10))
```

```python
# Selecting Required columns
events_oct.select("event_type", "product_id", "price").show(10)
```

```python
#products with price > 500
events_oct.filter("price > 500").count()
```

```python
# Grouping based on events
events_oct.groupBy("event_type").count().show()
```

```python
top_brands = events_oct.groupBy("brand").count().orderBy("count", ascending=False).limit(5)
display(top_brands)
```
